```
import requests
import urllib.request
import time
from bs4 import BeautifulSoup
import numpy as np
import pandas as pd
from urllib.request import urlopen
import json
y = {}
```

```
staturl = 'https://novelfull.com/warriors-promise/chapter-' 
i = 1
a = []
def urlgen(x):
  for i in range (1, x+1):
    url= staturl + str(i) + '.html'
    a.append(url)

urlgen(30)
#print(a)
```

```
user_agent = 'Mozilla/5.0 (Windows; U; Windows NT 5.1; en-US; rv:1.9.0.7) Gecko/2009021910 Firefox/3.0.7'
headers={'User-Agent':user_agent,} 
for i in range(0, len(a)):
  request=urllib.request.Request(a[i],None,headers) #The assembled request
  response = urllib.request.urlopen(request)
  data = response.read()
  soup = BeautifulSoup(data, "html")
  divs = soup.find("div", {"id": "chapter-content"})
  heading = divs.find_all('h3')
  myheading = heading[0].text
  index = myheading.split(':')[0]
  para = ''
  paragraphs = divs.find_all('p')
  for z  in paragraphs:
    text =  z.text
    text = text[42:]
    para+=text
  y[index] = myheading + para
#print(y['Chapter 2'])
```

```
json_object = json.dumps(y, indent = 4)   
print(json_object)  
```
